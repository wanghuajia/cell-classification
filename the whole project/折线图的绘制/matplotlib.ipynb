```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
data = pd .read_csv('./Xception/2018_08_10_log.csv')
acc=data['acc']
loss = data['loss']
val_loss = data['val_loss']
val_acc = data['val_acc']
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.plot(loss,color='r',linestyle='dashed',marker='o',label='loss')
plt.plot(val_loss,color='k',linestyle='dashed',marker='o',label='val_loss')
plt.legend(loc='best')
ax.set_title('The changing curve of Loss')
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
```
